# Estimativa de π por Série de Taylor

Este notebook foi desenvolvido para explorar a aproximação do número π utilizando a série de Taylor da função arco-tangente. A partir dessa série, implementei funções que permitem:

- Calcular aproximações de π usando um número limitado de termos.
- Gerar uma tabela com os valores aproximados e seus respectivos erros.
- Visualizar, por meio de gráficos, como o erro absoluto diminui à medida que aumentamos a quantidade de termos considerados.

Com isso, é possível analisar de forma prática o comportamento da série de Taylor aplicada ao cálculo de π, entendendo o impacto do truncamento da série e a relação entre precisão e número de termos.


```python
import math
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Função para calcular aproximação de pi com k termos
def pi_aprox(k: int) -> float:
  """
  Calcula a aproximaçao de pi usando k termos da série de Taylor de arctan(1).
  """
  return 4 * sum(((-1)**n) / (2*n + 1) for n in range(k))
```

```python
# Função para gerar tabela de aproximações
def gerar_tabela(ks):
  resultados = []
  for k in ks:
    aprox = pi_aprox(k)
    erro = abs(math.pi - aprox)
    resultados.append((k, aprox, erro))
  df = pd.DataFrame(resultados, columns=["k", "pi_k (aprox)", "Erro absoluto"])
  return df
```

```python
# Função para gerar gráfico do erro
def plotar_erro(ks):
  erros = [abs(math.pi - pi_aprox(k)) for k in ks]
  plt.figure(figsize=(8,5))
  plt.plot(ks, erros, marker='o')
  plt.xscale("log")
  plt.yscale("log")
  plt.xlabel("Número de termos (k)")
  plt.ylabel("Erro absoluto")
  plt.title("Erro na aproximação de pi por série de Taylor")
  plt.grid(True, which="both", linestyle="--", lw=0.5)
  plt.show()
```

```python
# Exemplo de uso

if __name__ == "__main__":
  ks = [1, 2, 3, 5, 10, 100, 1000, 10000]
  tabela = gerar_tabela(ks)
  print(tabela)
  plotar_erro(ks)
```
